## DATA ENCODING

Data encoding is the process of converting categorical data into numerical form so that machine learning algorithms can work efficiently.

### Types of Data Encoding
1. Nominal / One-Hot Encoding (OHE)  
2. Label / Ordinal Encoding  
3. Target-Guided Encoding  

---

## NOMINAL / ONE-HOT ENCODING (OHE)

One-Hot Encoding, also known as Nominal Encoding, is a technique used to represent categorical variables as numerical values.

### Definition
In One-Hot Encoding:
- Each unique category is represented as a binary vector  
- The vector length equals the number of unique categories  
- Only one element in the vector is `1`, and the rest are `0`  

This method is used when categorical data has **no inherent order**.

---

### Example

Consider a categorical feature **Color** with the following values:
- Red  
- Green  
- Blue  

After applying One-Hot Encoding:

| Color | Red | Green | Blue |
|------|-----|-------|------|
| Red  | 1   | 0     | 0    |
| Green| 0   | 1     | 0    |
| Blue | 0   | 0     | 1    |

---

### Vector Representation
- Red → `[1, 0, 0]`  
- Green → `[0, 1, 0]`  
- Blue → `[0, 0, 1]`  

---

### Key Points
- Used for nominal categorical variables  
- Avoids giving any ranking to categories  
- Can increase dimensionality when categories are many  


```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
## Create a single dataframe

df = pd.DataFrame({
    'color' : ['red','blue','green','green','red','blue']
})
```

```python
df.head()
```

```python
## Create a instance of OneHotEncoder
encoder = OneHotEncoder()
```

```python
## Perform Fit and Transform
encoded = encoder.fit_transform(df[['color']]).toarray()
```

```python
encoded_df = pd.DataFrame(encoded,columns = encoder.get_feature_names_out())
encoded_df.head()
```
