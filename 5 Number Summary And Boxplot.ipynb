### 5 NUMBER SUMMARY AND PLOT

#### Minimum, Maximum , Median , Q1 , Q3

```python
import numpy as np
```

```python
lst_marks = [45,32,56,75,89,54,32,89,98,87,67,54,45,98,99,67,74]
Minimum,Maximum,Q1,Median,Q3 = np.quantile(lst_marks,[0.0,1.0,0.25,0.50,0.75])
```

```python
print(Minimum,Maximum,Q1,Median,Q3)
```

```python
IQR = Q3 - Q1
print(IQR)
```

```python
lower_fence = Q1 - 1.5*(IQR)
higher_fence = Q3 + 1.5*(IQR)
```

```python
print(lower_fence,higher_fence)
```

```python
import seaborn as sns
sns.boxplot(lst_marks)
```
