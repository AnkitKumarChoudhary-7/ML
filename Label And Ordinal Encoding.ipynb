## LABEL ENCODING

Label Encoding (also known as Ordinal Encoding) is a technique used to convert categorical variables into numerical values.

### Definition
Label Encoding involves assigning a **unique numerical label** to each category in a categorical variable.  
The labels are usually assigned:
- In **alphabetical order**, or  
- Based on the **frequency** of the categories  

⚠️ This method may introduce an **artificial order** among categories.

---

### Example

Consider a categorical feature **Color** with the following values:
- Red  
- Green  
- Blue  

Using Label Encoding, the categories can be represented as:

| Color | Encoded Value |
|------|---------------|
| Red  | 1 |
| Green | 2 |
| Blue | 3 |

---

### Key Points
- Simple and memory efficient  
- Suitable when categories have an **ordinal relationship**  
- Not recommended for **nominal data** (can mislead models)



```python
import pandas as pd
```

```python
## Create a single dataframe
df = pd.DataFrame({
    'color' : ['red','blue','green','green','red','blue']
})
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

lbl_encoder = LabelEncoder()
```

```python
lbl_encoder.fit_transform(df[['color']])
```

```python
lbl_encoder.transform([['red']])
```

```python
lbl_encoder.transform([['blue']])
```

```python
lbl_encoder.transform([['green']])
```

## ORDINAL ENCODING

Ordinal Encoding is used to encode categorical variables that have an **intrinsic order or ranking**.

### Definition
In this technique, each category is assigned a **numerical value based on its relative order**.  
The numerical values reflect the ranking among categories, but not the exact difference between them.

---

### Example

Consider a categorical feature **Education Level** with the following ordered values:
- High School  
- College  
- Graduate  
- Post Graduate  

Using Ordinal Encoding, the representation will be:

| Education Level | Encoded Value |
|-----------------|---------------|
| High School     | 1 |
| College         | 2 |
| Graduate        | 3 |
| Post Graduate   | 4 |

---

### Key Points
- Suitable for features with **natural ordering**
- Preserves rank information  
- Should not be used for **nominal categories**
- The numeric gap does **not** represent actual magnitude



```python
## Ordinal Encoding 
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({
    'size': ['small','medium','large','medium','small','large']
})
```

```python
df
```

```python
## Create a instance of Ordinal Encoder and then fit transform
odl_encoder = OrdinalEncoder(categories = [['small','medium','large']])
```

```python
odl_encoder.fit_transform(df[['size']])
```

```python
odl_encoder.transform([['medium']])
```
