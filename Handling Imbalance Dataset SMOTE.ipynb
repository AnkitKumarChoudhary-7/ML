### SMOTE(Synthetic Minority Oversampling Technique)

##### SMOTE is a technique used in machine learning used to address imbalaced datasets where the minority classes has significantly fewer 
##### instances than the majority class.SMOTE involves generating synthetic instances of the minority class by interpolating between existing instances.

```python
from sklearn.datasets import make_classification
X,y = make_classification(n_samples = 1000,
                          n_features = 2,
                          n_clusters_per_class = 1,
                           weights = [0.90],
                          random_state = 12,
                          n_redundant = 0)
```

```python
X
```

```python
y
```

```python
import pandas as pd
df1 = pd.DataFrame(X,columns = ['f1','f2'])
df2 = pd.DataFrame(y,columns = ['target'])
final_df = pd.concat([df1,df2],axis = 1)
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'],final_df['f2'],c =final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
## Transform the dataset
oversample = SMOTE()
X,y = oversample.fit_resample(final_df[['f1','f2']],final_df['target'])
```

```python
df1 = pd.DataFrame(X,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
oversampled_df = pd.concat([df1,df2],axis = 1)
```

```python
plt.scatter(oversampled_df['f1'],oversampled_df['f2'],c = oversampled_df['target'])
```
