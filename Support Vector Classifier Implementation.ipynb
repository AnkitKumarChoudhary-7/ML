### Support Vector Classifier

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
### Lets Create data points
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=1,n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0],pd.DataFrame(X)[1]
```

```python
sns.scatterplot(x = pd.DataFrame(X)[0],y = pd.DataFrame(X)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state = 42)
```

```python
from sklearn.svm import SVC
```

### Using Linear Kernel

```python
svc = SVC(kernel = 'linear')
```

```python
svc.fit(X_train,y_train)
```

```python
svc.coef_
```

### Prediction

```python
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

### Using Poly Kernel

```python
poly = SVC(kernel = 'poly')
poly.fit(X_train,y_train)
y_pred = poly.predict(X_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

### Using RBF Kernel

```python
rbf = SVC(kernel = 'rbf')
rbf.fit(X_train,y_train)
y_pred = rbf.predict(X_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

### Using Sigmoid Kernel

```python
sigmoid = SVC(kernel = 'sigmoid')
sigmoid.fit(X_train,y_train)
y_pred = sigmoid.predict(X_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

### Using Precomputed Kernel

```python
precomputed = SVC(kernel = 'sigmoid')
precomputed.fit(X_train,y_train)
y_pred = precomputed.predict(X_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

### HyperParameter Tuning with SVC

```python
from sklearn.model_selection import GridSearchCV

## Defining Parameter Range
param_grid = {'C' :[0.1,1,10,100,1000],
             'gamma' : [1,0.1,0.01,0.001],
             'kernel' : ['rbf','linear','poly','sigmoid']}
```

```python
grid = GridSearchCV(SVC(),param_grid=param_grid,refit = True,cv=5,verbose = 3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred3 = grid.predict(X_test)
print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
```
