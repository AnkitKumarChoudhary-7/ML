## TARGET-GUIDED ORDINAL ENCODING

Target-Guided Ordinal Encoding is a technique used to encode categorical variables based on their **relationship with the target variable**.

### Definition
This encoding method is especially useful when:
- A categorical feature has a **large number of unique categories**
- The feature has a **strong influence on the target variable**

Instead of assigning arbitrary numbers, categories are encoded using **statistics derived from the target variable**.

---

### How It Works
In Target-Guided Ordinal Encoding:
- Each category is replaced with the **mean or median of the target variable**
- Categories are then ordered based on these target statistics
- This creates a **monotonic relationship** between the encoded feature and the target

---

### Example (Conceptual)

| Category | Mean Target Value | Encoded Value |
|---------|------------------|---------------|
| A       | 200              | 3 |
| B       | 150              | 2 |
| C       | 80               | 1 |

---

### Key Points
- Uses **target information** for encoding  
- Handles **high-cardinality categorical variables** efficiently  
- Can improve **predictive performance**  
- Risk of **data leakage** if not applied correctly  
- Should be fitted **only on training data**



```python
import pandas as pd

## Create a sample dataframe with ctaegorical variable and a target variable
df = pd.DataFrame({
    'city':['New York','London','Paris','Tokyo','New York','Paris'],
    'price': [200,150,300,250,180,320]
})
```

```python
df
```

```python
df.groupby('city')['price'].mean()
```

```python
mean_price = df.groupby('city')['price'].mean().to_dict()
mean_price
```

```python
df['city_encoded'] = df['city'].map(mean_price)
```

```python
df[['price','city_encoded']]
```

```python
import seaborn as sns
data = sns.load_dataset('tips')
data
```

```python
data.groupby('time')['total_bill'].mean()
```

```python
mean_total_bill = data.groupby('time')['total_bill'].mean().to_dict()
mean_total_bill
```

```python
data['time_encode'] = data['time'].map(mean_total_bill)
data[['total_bill','time_encode']]
```
